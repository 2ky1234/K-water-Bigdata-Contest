```python
# 관측소리스트로 반복문(전체데이터)
stnlist = stnIds_list
string = ''
for stn in stnlist:
    query = '?'+'apiKey='+apiKey+'&type='+type+'&dataCd='+dataCd+'&dateCd='+dateCd+'&startDt='+startDt+'&endDt='+endDt+'&stnIds='+str(stn)+'&schListCnt='+schListCnt+'&pageIndex='+pageIndex
    res = requests.get(url+query)
    string = string + res.text
gisang_raw = string
```

```python
# json파일 저장
gisang = json.loads(gisang_raw)
with open('gisang.json', 'w', encoding="utf-8") as make_file:
    json.dump(gisang, make_file, ensure_ascii=False, indent="\t")
```
